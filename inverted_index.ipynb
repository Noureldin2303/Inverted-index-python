```python
# Read the documents

docs = open('Documents.txt').read()

print(docs)

line = 1

for word in docs:
    if word == '\n':
        line += 1

print("\nNumber of documents: ",line)
```

```python
arr = []

for i in range(line):
    arr.append(docs.split('\n')[i])

arr
```

```python
# to maintain uniformity
docs = docs.lower()

print(docs) 
```

```python
# Tokenization

tokens = docs.split()

tokens
```

```python
# Stopping

stop_words = open('StopWords.txt', 'r').read()
stop_words = stop_words.split()

NewList = []
# Remove special characters
for doc in tokens:
    Newdoc = "".join(ch for ch in doc if ch.isalnum()) 
    NewList.append(Newdoc)

# Remove stopring words
after_stopping = [token for token in NewList if token not in stop_words]

after_stopping
```

```python
# Inverted Index

Inverted_index = {}

for token in after_stopping:
    for i in range(line):
        if token in arr[i]:
            if token in Inverted_index:
                Inverted_index[token].add(i+1)
            else:
                Inverted_index[token] = {i+1}


Inverted_index
```

```python
indexer = {}

for token in after_stopping:
    for i in range(line):
        if token in arr[i]:
            indexer[token] = i+1

indexer
```

```python
# Term squences

sorted(indexer.items(), key=lambda x: x[1])
```

```python
# Sorting Alphabetically

sorted(indexer.items(), key=lambda x: x[0])
```

```python
# Term Frequency

Term_frequancy = {}

for term in Inverted_index:
    Term_frequancy[term] = term
    Term_frequancy[term] = len(Inverted_index[term])

Term_frequancy
```
